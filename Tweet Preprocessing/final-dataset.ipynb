```python
dfcovid=pd.read_csv('../input/finaltweets/covidtweetsdataset.csv',index_col=[0])
dfcovid
```

```python
dfcovid2=pd.read_csv('../input/covidtweetsdataset2/finalcovidtweets2.csv',index_col=[0])
dfcovid2
```

```python
dfcovid=dfcovid.append(dfcovid2,ignore_index=True)
dfcovid
```

```python
dfcovid=dfcovid.drop_duplicates(subset='text')
dfcovid=dfcovid.reset_index(drop=True)
dfcovid
```

```python
dfcovid['text']=dfcovid['text'].apply(lambda x:x.replace("', '"," "))
dfcovid['text']=dfcovid['text'].apply(lambda x:x.replace("['",""))
dfcovid['text']=dfcovid['text'].apply(lambda x:x.replace("']",""))
dfcovid
```

```python
dfnoncovid=pd.read_csv('../input/finaltweets/sentiment140dataset.csv',index_col=[0])
dfnoncovid
```

```python
dfcovid=dfcovid.drop_duplicates(subset='text')
dfcovid=dfcovid.reset_index(drop=True)
dfnoncovid
```

```python
dfnoncovid['text']=dfnoncovid['text'].apply(lambda x:x.replace("', '"," "))
dfnoncovid['text']=dfnoncovid['text'].apply(lambda x:x.replace("['",""))
dfnoncovid['text']=dfnoncovid['text'].apply(lambda x:x.replace("']",""))
dfnoncovid
```

```python
dfcovid['label'] = pd.Series([1 for x in range(len(dfcovid.index))])
dfnoncovid['label'] = pd.Series([0 for x in range(len(dfnoncovid.index))])
```

```python
finaldf=pd.concat([dfcovid,dfnoncovid])
finaldf
```

```python
finaldf= finaldf.sample(frac=1).reset_index(drop=True)
finaldf
```
